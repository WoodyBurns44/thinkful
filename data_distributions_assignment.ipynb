```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1) Gamma Distribution #1-3

gammavar = np.random.gamma(4,scale=10, size=100)
mean = gammavar.mean()
std = gammavar.std()
```

```python
plt.hist(gammavar)
plt.axvline(x = mean, color= "b", linestyle = "solid", linewidth = 2)
plt.axvline(x= mean+std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.axvline(x= mean-std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.show()
```

```python
# 4. This distribution is skewed to the right (positive). It contains a lot of large outliers. 
```

```python
#2) Standard Exponential 1-3 
stan_expo_var = np.random.standard_exponential(size=100)
mean = stan_expo_var.mean()
std = stan_expo_var.std()
plt.hist(stan_expo_var)
plt.axvline(x = mean, color= "b", linestyle = "solid", linewidth = 2)
plt.axvline(x= mean+std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.axvline(x= mean-std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.show()
```

```python
# 4. The mean and standard deviation of standard exponential data are in line with
#    most of the data, which shows a right, positive skew with a few large outliers. 
```

```python
#3) Gumbel Distribution #1-3
gumbel_var = np.random.gumbel(size= 100)
mean = gumbel_var.mean()
std = gumbel_var.std()
plt.hist(gumbel_var)
plt.axvline(x = mean, color= "b", linestyle = "solid", linewidth = 2)
plt.axvline(x= mean+std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.axvline(x= mean-std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.show()
```

```python
# #4 the mean and standard deviation are in line with most of the data. However, the right positive skew 
#    shows some large outliers. 
```

```python
#4) ChiSquare Distribution 
chisquare_var = np.random.chisquare(df=2, size= 100)
mean = chisquare_var.mean()
std = chisquare_var.std()
plt.hist(chisquare_var)
plt.axvline(x = mean, color= "b", linestyle = "solid", linewidth = 2)
plt.axvline(x= mean+std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.axvline(x= mean-std, color= 'b', linestyle = "dashed", linewidth = 2)
plt.show()
```

```python
# 4. the chisqaure distribution is interesting because it is almost a perfectly negative linear 
#    relationship. The mean and standar deviation show a right skew. 
```

```python
#5) Multinomial Distribution 
multinomial_var = np.random.multinomial(20, [1/6.]*6, size=100)
mean = multinomial_var.mean()
std = multinomial_var.std()
plt.hist(multinomial_var)
plt.axvline(x= mean, color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(x= mean-std, color = 'b', linestyle = 'dashed', linewidth= 2)
plt.axvline(x=mean+std, color = 'b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
# 4. the mean and standard contain most of the data, with a slight right skew. 
```

```python
#6) Triangular distribution 
triangular = np.random.triangular(0, 10, 100, 100)
mean = triangular.mean()
std = triangular.std()
plt.hist(triangular)
plt.axvline(x=mean, color='b', linestyle= 'solid', linewidth= 2)
plt.axvline(x = mean-std, color = 'b', linestyle= 'dashed', linewidth= 2)
plt.axvline(x=mean+std, color='b', linestyle='dashed', linewidth= 2)
plt.show()
```

```python
# 4. the mean and standard deviation show a right skew in the data. 
```
